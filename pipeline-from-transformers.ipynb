```python
from transformers import pipeline

```

```python
# sentiment analysis figures out the sentiment of the text, either positive or negative
classifier = pipeline("sentiment-analysis")
sentences = ["This sentence may or may not contain hateful words",
            "Why do people love to be happy?",
            "I was delighted to hear the news from Sue"
           ]
classifier(sentences)
```

```python
# "ner" or Named Entity Recognition extracts the names in a sentence
classifier = pipeline("ner")
classifier(sentences)
```

```python
# feature-extraction returns the values of the hidden state vectors
classifier = pipeline("feature-extraction")
classifier("This is a sentence")
```

```python
# fill-mask fills a mask (LOL)
classifier = pipeline("fill-mask")
classifier("Lionel Messi plays in <mask>")
```

```python
# question-answering extracts an answer from a given passage
classifier = pipeline("question-answering")
context = r"""Extractive Question Answering is the task of extracting an answer from a text given a question. An example of a
 question answering dataset is the SQuAD dataset, which is entirely based on that task. If you would like to fine-tune
 a model on a SQuAD task, you may leverage the examples/pytorch/question-answering/run_squad.py script.
 """
output = classifier(question="What is extractive question answering?",context = context)
print(output)
output = classifier(question="What is a good example of a question answering dataset",context = context)
print(output)
```

```python
# text-generation generates sentences in incomplete strings
# any models can be used from the Hub by specifying the model in the parameters
classifier = pipeline('text-generation', model = 'distilgpt2')
classifier("In this course we will learn how to")
```

```python
# translation does what the word suggests, translates from one language to another
classifier = pipeline('translation_en_to_de')
print(classifier("What happened to the pickles in the jar?", max_length=30))
print(classifier("How do you turn potatoes into fries?", max_length=30))

classifier = pipeline('translation_en_to_fr')
print(classifier("What happened to the pickles in the jar?", max_length=30))
print(classifier("How do you turn potatoes into fries?", max_length=30))
```

```python
# zero shot classification helps to specify which labels to use for the classification for strings with no labels
classifier = pipeline('zero-shot-classification')
classifier("This sentence is about the Transformers library", candidate_labels=["education","politics","sports"])
```
